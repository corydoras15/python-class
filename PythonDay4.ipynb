“알고리즘은 문제 해결을 위한 절차를 하나하나 구체적인 순서에 따라 표현한 아이디어 또는 생각”

알고리즘은 비단 컴퓨터에서만 사용되는 언어가 아니라 알고리즘 자체는 일상생활에서도 많이 사용된다. 예를 들면, 요리의 레시피, 음악의 악보, 가전제품 설명서 등등 이 모두 알고리즘의 예이다. 

그 알고리즘을 표현하기 위해 문자, 기호, 그림등으로 나타낼수 있다. 

이러한 알고리즘을 프로그래밍 언어로 기술하게 되면 그것이 바로 프로그램이 된다. 

프로그래밍 언어로 작성한 프로그램은 컴퓨터에 지시하기 위한 언어이다.
